```python
import numpy as np
import pandas as pd
df=pd.DataFrame({
    'length':[1.5,0.5,1.2,0.9,3],
    'width':[0.7,0.2,0.15,0.2,1.1]
}, index=['pig','rabbit','duck','chicken','horse'])
hist=df.hist(bins=3)
```

```python
import matplotlib.pyplot as plt
d=np.random.uniform(0,40,100)
n, bins, patches=plt.hist(x=d, bins='auto', color='#0504aa', alpha=0.7, rwidth=0.85)
plt.grid(axis='y',alpha=0.75)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('My very own histogram')
plt.text(23, 45, r'$\mu=15, \sigma=3$')
maxfreq=n.max()
plt.ylim(ymax=np.ceil(maxfreq/10)*10 if maxfreq % 10 else maxfreq+10)
```

```python
import matplotlib.pyplot as plt
import numpy as np
duration = np.random.uniform(60, 300, 100)  

plt.hist(x=duration, bins='auto', color='#007acc', alpha=0.7, rwidth=0.85)

plt.xlabel('Duration (seconds)')
plt.ylabel('Frequency')
plt.title('Old Faithful Eruption Duration (Seconds)')

plt.grid(axis='y', alpha=0.75)

plt.show()
```

```python
data = pd.read_excel('15_OLDFAITH.xls')

# Create a DataFrame
df = pd.DataFrame(data)

# Plot histogram for 'Duration' (Old Faithful Eruption Duration)
plt.hist(df['Duration'], bins='auto', color='#0504aa', alpha=0.7, rwidth=0.85)
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Duration (giây)')
plt.ylabel('Frequency')
plt.title('Histogram of Old Faithful Eruption Duration (Dataset 15)')
plt.show()
```

```python
diet_coke_weight = np.random.uniform(0.5, 2.5, 100)
plt.hist(x=diet_coke_weight, bins='auto', color='#ff5733', alpha=0.7, rwidth=0.85)
plt.xlabel('Weight (pounds)')
plt.ylabel('Frequency')
plt.title('Diet Coke Weight Distribution (Pounds)')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
